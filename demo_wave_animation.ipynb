```python
%matplotlib nbagg
# require matplotlib ver. 1.4+
#  Option 'nbagg' enables interactive figures

from matplotlib import pyplot as plt
from matplotlib import animation
import numpy as np
import pandas as pd


#set up the figure
fig = plt.figure(figsize=(8,3))
ax1=fig.add_subplot(1,2,1)
ax2=fig.add_subplot(1,2,2)

# x values
x = np.arange(0, 4*np.pi, 0.1)

#initialization of animation
def my_init():
    return ax1, ax2

#plot sub-figures
def my_animate(i):
    ax1.cla()
    ax1.plot(x,np.sin(x + i/10), "b")
    ax1.set_xlim([0, x[-1]])
    ax1.set_ylim([-1.2,1.2])
    ax1.set_title('Sin function')
    ax1.grid()
    
    ax2.cla()
    ax2.plot(x,np.cos(x + i/10), "r")
    ax2.set_xlim([0, x[-1]])
    ax2.set_ylim([-1.2,1.2])
    ax2.set_title('Cos function')
    ax2.grid()

    return ax1, ax2

# generating frames (100 frames)
anim = animation.FuncAnimation(fig, my_animate, init_func=my_init, repeat = False, frames=100, interval=200, blit=False)
plt.show()

# To output a movie into a file, run
#
# anim.save('sin_cos.mp4')
#
# instead of plt.show(). 
#
# The save function uses 'ffmpeg' (https://ffmpeg.org/).
# It was download it and put into /usr/local/bin.
```
